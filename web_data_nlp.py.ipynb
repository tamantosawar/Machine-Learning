```python
# web_data_nlp
import urllib.request
from bs4 import BeautifulSoup
website="http://php.net"
#data downloading 
web_data=urllib.request.urlopen(website)
#printing data
#print(web_data.read())
#reading data with html tags
clean_data=web_data.read()
#applying library of html 5 to scrape
get_clean=BeautifulSoup(clean_data,'html5lib')
#getting only text format data
final_data=get_clean.get_text()
good_data=final_data.strip()

new_data=[]
for i in good_data:
    j=i.split()
    new_data.append(j)
print(new_data)
```

```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
data="hello every one , I am Tamanna Soni to be a"
y=word_tokenize(data)
print(y)
for i in y:
    if i in stopwords.words('english'):
      y.remove(i)
print(y)  

```
