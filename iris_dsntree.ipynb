```python
#using decesion algo
#!/usr/bin/python
import numpy
from sklearn import tree
from sklearn.datasets import load_iris

#loading all data
iris=load_iris()  
#printing features name
#print(iris.feature_names)

#printing target name 
print(iris.target_names)

x=[0,50,100]
only_target_training=numpy.delete(iris.target,x)
only_data_training=numpy.delete(iris.data,x,axis=0)
#print(only_target_training)
print(only_data_training)
print(only_target_training.size)

#training data
#features data
#print(iris.data)

#only setosa_data
#setosa=iris.data[0:50]
#print(setosa)
#
#targetdata means flowers data
#print(iris.target)
#s_data=iris.target[0:50]
#print(s_data)

#testing target
test_data=iris.data[x]
test_target=iris.target[x]
print(test_data)
print(test_target)

#calling algo

clf=tree.DecisionTreeClassifier()
trained=clf.fit(only_data_training, only_target_training)
output=trained.predict(test_data)
print(output)
```

```python
#using decesion algo
#!/usr/bin/python
import numpy
from sklearn import tree
from sklearn.datasets import load_iris
#loading all data
iris=load_iris()  

#now splitting into test &train datasets
from sklearn.model_selection import train_test_split

x,y,z,a=train_test_split(iris.data,iris.target,test_size=0.1)
'''here
x is train_iris{all features values containing 90% }
y is remaining test_iris{10% of features }
z is train_target{all label containing 90% of iris.target}
a is  test_target {10% of iris.target }
'''

#calling algo

clf=tree.DecisionTreeClassifier()
# trained data with decision tree

trained=clf.fit(x,z)
#pridiction
output=trained.predict(y)
print(output)
print(a)
#print(y)

from sklearn.metrics import accuracy_score
check_pct= accuracy_score(a,output)
print(check_pct)
```
